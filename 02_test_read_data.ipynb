```python
import pandas as pd
from _02_preprocessing.preprocessing import data_1_json2df, data_2_json2df, data_3_json2df
```

```python
df=pd.read_csv('raw_data/StationCode_mapper.csv',index_col=0)

subwayline = '2호선'
stationname = '강남역'
dow = 'MON'
hh = '08'

station_code=df[(df['subwayLine']==subwayline) & (df['stationName']==stationname)]['stationCode'].values[0]
```

```python
def set_time(df):
    df['time'] = df['hh']+df['mm']
    df['time']=pd.to_datetime(df['time'],format='%H%M')
    return df
```

```python
df_1 = data_1_json2df(station_code,dow,hh)
df_1['congestionTrain']=df_1['congestionTrain'].apply(lambda x: x if x!=0 else None)
df_1.dropna(inplace=True)
df_1=set_time(df_1)

df_2 = data_2_json2df(station_code,dow,hh)
for i in range(len(df_2['congestionCar'][0])):
    df_2[f'혼잡도_{i+1}번칸'] = df_2['congestionCar'].apply(lambda x : x[i])
df_2['congestionCar']=df_2['congestionCar'].apply(lambda x: sum(x) if sum(x) !=0 else None)
df_2.dropna(inplace=True)
df_2=set_time(df_2)


df_3 = data_3_json2df(station_code,dow,hh)
for i in range(len(df_3['getOffCarRate'][0])):
    df_3[f'하차수_{i+1}번칸'] = df_3['getOffCarRate'].apply(lambda x : x[i])
df_3['getOffCarRate']=df_3['getOffCarRate'].apply(lambda x: sum(x) if sum(x) !=0 else None)
df_3.dropna(inplace=True)
df_3=set_time(df_3)
```
